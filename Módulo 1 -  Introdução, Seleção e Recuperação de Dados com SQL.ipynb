## Selecionando e recuperando dados com SQL

### O que é SQL afinal?

Esta é a **Linguagem de Consulta Estruturada**.

Esta é a linguagem padrão para muitos sistemas de gerenciamento de banco de dados relacional e manipulação de dados. **SQL é usado frequentemente para consultar, inserir, atualizar e modificar dados**. Em um nível básico SQL é um método para comunicação entre você e o banco de dados. 

Uma das grandes coisas sobre SQL é que ele é composto de declarações que são **palavras descritivas**. Em outras palavras, muitos dos comandos usados no SQL são bastante fáceis de interpretar em comparação com muitas outras linguagens de computador. Isso torna o SQL, como linguagem, muito fácil de entender e aprender. No entanto, é importante entender que **SQL é uma linguagem não processual**. Isso significa que você não será capaz de escrever aplicativos completos com ele, mas o que você pode fazer é interagir e se comunicar com dados. Isso torna relativamente simples, mas também linguagem muito poderosa. 

Quando você pensa em SQL, tudo o que você precisa pensar são dados. **SQL é tudo sobre dados**. SQL é realmente usado para três coisas. 

1. É usado para ler e recuperar dados, portanto, dados geralmente são armazenados em um banco de dados, e você deseja recuperá-los ou lê-los. E você pode usar o SQL como um meio de ser um tradutor para isso. 

2. QL também é usado como uma maneira de escrever dados em um banco de dados. Então, se você precisa escrever dados em uma tabela ou inserir novos dados, você pode usar SQL como um meio para fazer isso. 

3. E, finalmente, ele é usado para atualizar e inserir novos dados. 

Como você pode ver, SQL tem um design realmente simples, certo. Está muito contido no que é capaz de fazer, que é **ler, escrever e atualizar dados**. Devido a isso você vai descobrir que há um monte de pessoas que são capazes de usar esta linguagem.

Se olharmos para este gráfico, o que vemos é a linguagem SQL classificada pelo número de trabalhos de programação. Este é o formulário Indeed.com em 2016. Ele classificou SQL como o idioma número um. Há muitos trabalhos lá fora que exigem o uso do SQL, e não é apenas para ciência de dados. 



É importante entender como outras pessoas usam SQL e como outras pessoas além de cientistas de dados e programadores podem estar usando isso. Há muitas, muitas pessoas que podem usar SQL em seus trabalhos. Isso inclui tudo, desde desenvolvedores de back-end, engenheiros de controle de qualidade, arquitetos de dados, engenheiros de sistema , obviamente cientistas de dados e até analistas de dados. 

Vamos falar um pouco mais sobre os **DBAs**, ou os **administradores de banco de dados** , e como eles se comparam com o cientista de dados. Um ***DBA é responsável por gerenciar todo o banco de dados e protegê-lo***. Um cientista de dados, por outro lado, é tipicamente um usuário desse banco de dados. O DBA será responsável por dar permissões às pessoas e determinar quem tem acesso a quais dados. Eles geralmente são responsáveis por gerenciar as tabelas e criá-las. No entanto, isso é algo que você provavelmente terá que obter os direitos frequentemente de um DBA. 

As formas como as duas posições, DBA e cientista de dados, são similares é que ambos usam SQL para entender os dados, consulta-los e recuperá-los. Ambos escrevem consultas muito complexas, mas a principal diferença é que o cientista de dados é realmente o usuário final. Considerando que o DBA é aquele que o administra, o governa e gerencia o banco de dados, como um todo. Os cientistas de dados precisam ser capazes de recuperar dados. Sabemos que não podemos fazer nada até termos os dados para trabalhar , certo. Precisamos de uma maneira de ir e obter esses dados. 

SQL é realmente fundamental na ciência de dados porque você realmente não pode começar construindo modelos ou fazendo previsões até que você tenha os dados. SQL é o meio de entrar em um banco de dados e obter esses dados. Os cientistas de dados também podem usar isso para criar seu próprio ambiente de teste de tabela ou digamos que você construiu um modelo e deseja implantá-lo, e que deseja adicioná-lo de volta à tabela. Talvez seja necessário criar sua própria tabela ou ambiente de teste para adicioná-lo ao. Uma coisa que não é única com o cientista de dados, ou outras pessoas usando SQL, é que muitas vezes você está combinando várias tabelas juntas e um monte de vezes isso leva a consultas um pouco mais complexas para serem escritas para análise. Cientistas de dados, embora a maneira número um que eles estão usando SQL é realmente para ser capaz de recuperar seus dados para análise , eles podem fazer um pouco da análise usando SQL. No entanto, **a principal coisa para a qual eles estão usando SQL é para recuperação de dados**. 

A última coisa que quero salientar é que só porque você está aprendendo SQL na classe, a sintaxe do que você está escrevendo pode mudar um pouco com base no sistema de gerenciamento de banco de dados relacional que você está usando. Novamente, você pode pensar em SQL como o intérprete entre você e o banco de dados. Como você escreve parte da sintaxe para SQL vai depender do sistema de gerenciamento de banco de dados relacional com o qual você está interagindo. Estendendo nossa analogia você pode pensar nisso como o sotaque ou talvez como o dialeto. SQL é capaz de traduzi-lo para você, mas às vezes você tem que ajustar ele um pouco com base no sistema de gerenciamento de banco de dados que você está usando.
