## Quais so os tipos de tcnicas de modelagem de dados?

Tcnicas de modelagem de dados so os **diferentes mtodos que voc pode usar para criar modelos de dados distintos**. As abordagens evoluram ao longo do tempo como resultado de inovaes em conceitos de banco de dados e governana de dados. Veja a seguir os principais tipos de modelagem de dados:


### Modelagem de dados hierrquica

Na modelagem de dados hierrquica, voc pode **representar as relaes entre os vrios elementos de dados em um formato semelhante a uma rvore**. Modelos de dados hierrquicos representam relaes de um para muitos, com pais ou classes de dados raiz mapeados para vrios filhos.

No exemplo da concessionria de automveis, a classe principais Showrooms teria as entidades Carros e Vendedores como secundrios, porque h vrios carros e vendedores trabalhando em um showroom.


### Modelagem de dados de grafos

A modelagem de dados hierrquica evoluiu com o passar do tempo e se transformou na modelagem de dados de grafos. Modelos de dados de grafos **representam relaes de dados que tratam entidades igualmente**. As entidades podem se vincular entre si em relaes de um para muitos ou de muitos para muitos sem qualquer conceito de pai ou filho.

Por exemplo, um showroom pode ter vrios vendedores, e um vendedor tambm pode trabalhar em vrios showrooms se os seus turnos variarem de acordo com o local.


### Modelagem de dados relacional

A modelagem de dados relacional ** uma abordagem de modelagem popular que visualiza classes de dados como tabelas**. Diferentes tabelas de dados se unem ou se vinculam usando chaves que representam a relao de entidades do mundo real. Voc pode usar a tecnologia de banco de dados relacional para **armazenar dados estruturados, e um modelo de dados relacional  um mtodo til para representar sua estrutura de banco de dados relacional**.

Por exemplo, a concessionria de automveis teria modelos de dados relacionais que representam a tabela Vendedores e a tabela Carros, conforme mostrado aqui:


    ID do vendedor  Nome  

       1	        Jane
       
       2	        John
       
    ID do carro   Marca do carro
    
      C1	         XYZ
      
      C2	         ABC


**ID do vendedor e ID do carro so chaves primrias que identificam exclusivamente entidades individuais do mundo real**. Na tabela Showroom, essas chaves primrias atuam como chaves estrangeiras que vinculam os segmentos de dados.
 

    ID do showroom	Nome do showroom	ID do vendedor	ID do carro*

Em bancos de dados relacionais, **as chaves primria e estrangeira trabalham juntas para mostrar a relao de dados**. A tabela anterior demonstra que showrooms podem ter vendedores e carros.


### Modelagem de dados de relaes entre entidades

A modelagem de dados de relaes entre entidades (ER) **usa diagramas formais para representar as relaes entre entidades em um banco de dados**. Arquitetos de dados usam vrias ferramentas de modelagem de ER para representar dados.


### Modelagem de dados orientada a objetos

A programao orientada a objetos **usa estruturas de dados, chamadas de objetos, para armazenar dados**. Esses objetos de dados so abstraes de software de entidades do mundo real. Por exemplo, em um modelo de dados orientado a objetos, a concessionria de automveis teria objetos de dados, como Clientes, com atributos do tipo nome, endereo e nmero de telefone. Voc armazenaria os dados do cliente de modo que cada cliente do mundo real fosse representado como um objeto de dados de cliente.

### Modelagem de dados dimensional

A computao empresarial moderna usa a tecnologia de data warehouse para armazenar grandes quantidades de dados para anlise. Voc pode usar projetos de modelagem de dados dimensionais para **rmazenamento e recuperao de dados em alta velocidade a partir de um data warehouse**. Modelos dimensionais usam dados duplicados ou redundantes e priorizam a performance ao uso de menos espao para o armazenamento de dados.

Por exemplo, em modelos de dados dimensionais, a concessionria de automveis tem dimenses como Carro, Showroom e Tempo. A dimenso Carro tem atributos como nome e marca, mas a dimenso Showroom tem hierarquias como estado, cidade, nome da rua e nome do showroom.
