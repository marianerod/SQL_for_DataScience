## Selecionando e recuperando dados com SQL

### O que é SQL afinal?

Esta é a **Linguagem de Consulta Estruturada**.

   Esta é a linguagem padrão para muitos sistemas de gerenciamento de banco de dados relacional e manipulação de dados. **SQL é usado frequentemente para consultar, inserir, atualizar e modificar dados**. Em um nível básico SQL é um método para comunicação entre você e o banco de dados. 

   Uma das grandes coisas sobre SQL é que ele é composto de declarações que são **palavras descritivas**. Em outras palavras, muitos dos comandos usados no SQL são bastante fáceis de interpretar em comparação com muitas outras linguagens de computador. Isso torna o SQL, como linguagem, muito fácil de entender e aprender. No entanto, é importante entender que **SQL é uma linguagem não processual**. Isso significa que você não será capaz de escrever aplicativos completos com ele, mas o que você pode fazer é interagir e se comunicar com dados. Isso torna relativamente simples, mas também linguagem muito poderosa. 

Quando você pensa em SQL, tudo o que você precisa pensar são dados. **SQL é tudo sobre dados**. SQL é realmente usado para três coisas. 

1. É usado para ler e recuperar dados, portanto, dados geralmente são armazenados em um banco de dados, e você deseja recuperá-los ou lê-los. E você pode usar o SQL como um meio de ser um tradutor para isso. 

2. QL também é usado como uma maneira de escrever dados em um banco de dados. Então, se você precisa escrever dados em uma tabela ou inserir novos dados, você pode usar SQL como um meio para fazer isso. 

3. E, finalmente, ele é usado para atualizar e inserir novos dados. 

Como você pode ver, SQL tem um design realmente simples, certo. Está muito contido no que é capaz de fazer, que é **ler, escrever e atualizar dados**. Devido a isso você vai descobrir que há um monte de pessoas que são capazes de usar esta linguagem.

## Quais são os tipos de técnicas de modelagem de dados?

Técnicas de modelagem de dados são os **diferentes métodos que você pode usar para criar modelos de dados distintos**. As abordagens evoluíram ao longo do tempo como resultado de inovações em conceitos de banco de dados e governança de dados. Veja a seguir os principais tipos de modelagem de dados:


### Modelagem de dados hierárquica

Na modelagem de dados hierárquica, você pode **representar as relações entre os vários elementos de dados em um formato semelhante a uma árvore**. Modelos de dados hierárquicos representam relações de um para muitos, com pais ou classes de dados raiz mapeados para vários filhos.

No exemplo da concessionária de automóveis, a classe principais Showrooms teria as entidades Carros e Vendedores como secundários, porque há vários carros e vendedores trabalhando em um showroom.


### Modelagem de dados de grafos

A modelagem de dados hierárquica evoluiu com o passar do tempo e se transformou na modelagem de dados de grafos. Modelos de dados de grafos **representam relações de dados que tratam entidades igualmente**. As entidades podem se vincular entre si em relações de um para muitos ou de muitos para muitos sem qualquer conceito de pai ou filho.

Por exemplo, um showroom pode ter vários vendedores, e um vendedor também pode trabalhar em vários showrooms se os seus turnos variarem de acordo com o local.


### Modelagem de dados relacional

A modelagem de dados relacional **é uma abordagem de modelagem popular que visualiza classes de dados como tabelas**. Diferentes tabelas de dados se unem ou se vinculam usando chaves que representam a relação de entidades do mundo real. Você pode usar a tecnologia de banco de dados relacional para **armazenar dados estruturados, e um modelo de dados relacional é um método útil para representar sua estrutura de banco de dados relacional**.

Por exemplo, a concessionária de automóveis teria modelos de dados relacionais que representam a tabela Vendedores e a tabela Carros, conforme mostrado aqui:


    ID do vendedor  Nome  

       1	        Jane
       
       2	        John
       
    ID do carro   Marca do carro
    
      C1	         XYZ
      
      C2	         ABC


**ID do vendedor e ID do carro são chaves primárias que identificam exclusivamente entidades individuais do mundo real**. Na tabela Showroom, essas chaves primárias atuam como chaves estrangeiras que vinculam os segmentos de dados.
 

    ID do showroom	Nome do showroom	ID do vendedor	ID do carro*
    
       S1	         Showroom de NY	          1	            C1
 

Em bancos de dados relacionais, **as chaves primária e estrangeira trabalham juntas para mostrar a relação de dados**. A tabela anterior demonstra que showrooms podem ter vendedores e carros.


### Modelagem de dados de relações entre entidades

A modelagem de dados de relações entre entidades (ER) **usa diagramas formais para representar as relações entre entidades em um banco de dados**. Arquitetos de dados usam várias ferramentas de modelagem de ER para representar dados.


### Modelagem de dados orientada a objetos

A programação orientada a objetos **usa estruturas de dados, chamadas de objetos, para armazenar dados**. Esses objetos de dados são abstrações de software de entidades do mundo real. Por exemplo, em um modelo de dados orientado a objetos, a concessionária de automóveis teria objetos de dados, como Clientes, com atributos do tipo nome, endereço e número de telefone. Você armazenaria os dados do cliente de modo que cada cliente do mundo real fosse representado como um objeto de dados de cliente.

<div class="alert alert-info">
     
Modelos de dados orientados a objetos superam muitas das limitações dos modelos de dados relacionais e são populares em bancos de dados multimídia.
     
</div>


### Modelagem de dados dimensional

A computação empresarial moderna usa a tecnologia de data warehouse para armazenar grandes quantidades de dados para análise. Você pode usar projetos de modelagem de dados dimensionais para **rmazenamento e recuperação de dados em alta velocidade a partir de um data warehouse**. Modelos dimensionais usam dados duplicados ou redundantes e priorizam a performance ao uso de menos espaço para o armazenamento de dados.

Por exemplo, em modelos de dados dimensionais, a concessionária de automóveis tem dimensões como Carro, Showroom e Tempo. A dimensão Carro tem atributos como nome e marca, mas a dimensão Showroom tem hierarquias como estado, cidade, nome da rua e nome do showroom.

### Data Models, Part 3: Relational vs. Transactional Models

Agora vamos falar sobre o tipo de modelos de dados com os quais trabalharemos nesta aula. Após essa aula, você deve ser capaz de **definir e descrever modelos de banco de dados relacionais e transacionais, definir entidades, atributos e relacionamento, descrever e explicar a diferença entre um para um, um-para-muitos e muitos-para-muitos, descrever o uso de uma chave primária em um banco de dados e explicar como diagrama ER é usado para documentar e ilustrar relacionamentos**. 

Vamos começar falando um pouco sobre **modelos de banco de dados relacionais versus transacionais**. 

1. Um modelo relacional é um design de banco de dados que mostra as relações entre as diferentes tabelas e isso é realmente usado para otimizar dados de consulta, tornando fácil e intuitivo para acessar os dados. 

2. Transacional, por outro lado, você pode pensar como um banco de dados mais operacional. 

Se você estiver no setor de saúde, por exemplo, você pode ter um banco de dados transacional que é usado para armazenar todas as informações de reivindicações e, em seguida, essas informações podem não ser armazenadas de forma ótima para consultar e usá-las para análise. Na verdade, talvez você precise pegar e extrair essas informações transacionais do banco de dados e movê-la para um modelo relacional. A maior parte do que trabalharemos nesta aula é o modelo relacional. 

Os blocos de construção deste modelo relacional são realmente três coisas simples. Temos entidades que são uma pessoa, lugar, coisa ou evento. Estes são muito distinguíveis. Eles são únicos. Eles são distintos. Por exemplo, eu poderia ser uma entidade, Sadie St. Lawrence, e então temos atributos, que são características desta entidade. Como uma entidade, seria eu mesmo e, em seguida, um atributo sobre mim seria que eu sou mulher. Então, o terceiro bloco de construção do modelo é o relacionamento. Descreve a associação entre diferentes entidades. Há alguns tipos de relacionamentos em um banco de dados e os que eu quero cobrir são os três mostrados aqui: **um-para-muitos, muitos-para-muitos e um-para-um**. 

1. Se você pensar em um relacionamento **um-para-muitos**, este pode ser um cliente que tem muitas faturas. 

2. Quando você pensa em um relacionamento **muitos-para-muitos**, isso pode ser um exemplo de muitos alunos para muitas classes. Você pode ter um aluno que pertence a muitas classes diferentes ou você pode ter uma classe que tem muitos alunos diferentes. 

3. Então, se você pensar em um relacionamento **um-para-um**, este é um gerente de uma loja. Digamos que você tem uma loja de artigos esportivos e cada uma dessas lojas tem apenas um gerente. Isso seria um exemplo de uma relação um-para-um. 

Para entender essas relações entre as tabelas muito melhor, o que muitas vezes é usado para descrever isso são **diagramas ER**. Um modelo ER, em seguida, é composto por tipos de entidade e as relações específicas que podem existir entre essas entidades. Estes são geralmente exibidos em um formato visual e um relacionamento representa uma relação entre as tabelas. Ele muitas vezes ajuda você a entender e representar um processo de negócios e ele irá mostrar os links entre essas tabelas. Os links são realmente importantes porque em uma lição posterior, vamos aprender como juntar essas tabelas e combinar os dados. Ser capaz de olhar para este diagrama e ver como eles se relacionam entre si é realmente importante. 





O que usaremos para juntar essas tabelas juntas são duas coisas. Podemos usar as **chaves primárias ou chaves estrangeiras**. 

1. A **chave primária** é uma coluna ou conjunto de colunas cujos valores identificam exclusivamente cada linha em uma tabela. Então, isso nos permite pegar essas identidades únicas e, em seguida, juntar-se a outra tabela. 

2. **Chaves estrangeiras** são semelhantes, mas no que diz respeito a uma ou mais colunas podem ser usadas juntas para identificar uma única linha na tabela. 

Quando estamos olhando para diagramas de ER, que novamente é uma das maneiras que você vai começar a pensar antes de fazer, você vai olhar para talvez um diagrama de ER e entender quais elementos de dados você está tentando se juntar e como você precisa obtê-los. Mas uma das coisas que você precisa entender é como ler isso. Falamos um pouco sobre relacionamentos e as diferentes relações entre uma tabela. E então, havia um tipo diferente de notação que explica as relações. Temos a **notação de Chen**, e há a notação de **pé de corvo**, e depois há a notação de diagrama de **classe UML**. 

1. A notação Chen usa One-to-m para um relacionamento um-para-muitos, e M e N para um relacionamento muitos-para-muitos e um para um relacionamento um-para-um. 

2. Na notação de pé de corvo, temos os trilhos de trem que representam 1 e, em seguida, o pé de corvo que representa muitos. 

3. E finalmente, com UML temos um 1.1 que representa o conceito de um e 1, que representa o conceito de muitos. 

Definitivamente tomar notas sobre estes. Você vai estar olhando para diagramas ER com bastante frequência e você vai precisar entender essas notações ao ler diagramas ER para que você possa entender como você vai escrever sua consulta e juntar-se a tabela juntos ou até mesmo para descobrir o que está listado na tabela. Isso é realmente parte do pensamento antes de você começar a fazer o conceito que eu estava falando antes. **Ter uma boa compreensão de por que os dados são estruturados de uma maneira particular e como ler os diagramas ER será muito útil para você escrever suas consultas e garantir que você obtenha resultados precisos.**
