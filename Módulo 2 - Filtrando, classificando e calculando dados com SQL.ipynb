### Porqu ordernar dados?

Muitas vezes quando voc est olhando para dados, voc no vai ser capaz de olhar para todos os registros. **Classificar seus dados de forma lgica pode ajudar voc ver facilmente as informaes que voc deseja no topo**.  realmente til ser especfico sempre sobre seus dados que voc est recuperando, mas tambm, sobre como voc deseja fazer isso. 



> **ORDER BY permite classificar dados por colunas especficas**. 

Porm, existem algumas regras ao usar ORDER BY. 

- Ele pode tomar vrios nomes de coluna;
- Voc pode ordenar por uma coluna ou pode ordenar por todas as colunas, e assim vai da forma que voc deseja adicion-las. Se voc estiver fazendo vrias colunas, voc s quer ter certeza de que est adicionando uma vrgula depois disso;
- Voc pode classificar por uma coluna que voc no recuperou;
- Pode no estar em sua instruo SELECT, mas voc ainda pode usar a coluna para classificar seus dados, o que  realmente til; 
- ORDER BY deve ser sempre a ltima clusula na instruo SELECT. 

Apenas uma espcie de toque final, sempre arredondar para cima com a adio do ORDER BY no final. Ento, para fazer isso, voc pode classificar por posio da coluna. 



Aqui no final da minha consulta, eu tenho ordem por **colunas dois e trs na tabela**, ou voc pode at mesmo classific-lo pelos nomes reais da coluna. Eu costumo utilizar os nomes das colunas, e isso mantm as coisas realmente consistentes. H tambm algumas direes como com qualquer tipo de classificao, para que voc possa classificar tanto em **ordem ascendente, ASC**, ou **decrescente, DESC**. E ento, isso s  aplicado ao nome da coluna que prossegue diretamente. Se voc estiver usando ordem decrescente e tiver preo unitrio, ele no vai fazer isso para todas as colunas aps usar DESC. Voc tem que especificar cada coluna individual para crescente e decrescente, se voc quiser que seja assim. Ento, a idia de ordenao  um conceito bastante simples. Mas como voc pode ver,  realmente til, e apenas certificando-se de que quando voc est recuperando seus dados, voc est visualizando como voc quer ver, e realmente apenas chegando ao ncleo desses resultados. ## Grouping Data with SQL

Anteriormente, analisamos o uso de funes agregadas, como mdia, mnima, mxima e soma. No entanto, uma das coisas que no resolvemos  **em que voc est realmente agregando essas funes**. Algumas coisas importantes a serem observadas sobre o GROUP BYs  que **podemos agrup-lo por vrias colunas**. Podemos ter mais colunas do que apenas regies. Por exemplo, talvez queiramos saber regio, cidade, pas. Podemos listar todos eles em nossa clusula GROUP BY. Vamos apenas adicionar uma vrgula aps cada coluna que queremos listar. Ambos tm propsitos diferentes. Apenas certifique-se de anotar quando cada um  apropriado, para que voc no perca nenhum dado que deva ser contado. Outra coisa a ser observada ao trabalhar com GROUP BYs  que ** sempre uma boa prtica usar a clusula de ORDER BY**.
